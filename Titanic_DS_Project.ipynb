```python
# ILOC - uses numbers (position)
df.iloc[0:5, 0:3]    # First 5 rows, first 3 columns

# LOC - uses labels (names)
df.loc[0:5, ['Name', 'Age', 'Sex']]  # Rows 0-5, specific columns
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
```

```python
df['IsAlone'] = (df['FamilySize'] == 1).astype(int)
```

```python
df['Title'] = df['Name'].apply(lambda x: x.split(',')[1].split('.')[0].strip())
```

```python
df['AgeGroup'] = pd.cut(df['Age'], bins=[0,12,18,35,60,100], labels=['Child','Teen','Young','Adult','Senior'])
```

```python
# Age Group:

def age_group(age):
    if age <= 12:
        return 'Child'
    elif age <= 18:
        return 'Teen'
    elif age <= 35:
        return 'Young Adult'
    elif age <= 60:
        return 'Adult'
    else:
        return 'Senior'

df['AgeGroup'] = df['Age'].apply(age_group)
print("\n5. AGE GROUP:")
print(df[['Age', 'AgeGroup']].head(10))
```

```python
#fare group

def fare_group(fare):
    if fare <= 10:
        return 'Low'
    elif fare <= 50:
        return 'Medium'
    elif fare <= 100:
        return 'High'
    else:
        return 'Very High'

df['FareGroup'] = df['Fare'].apply(fare_group)
print("\n6. FARE GROUP:")
print(df[['Fare', 'FareGroup']].head(10))
```
